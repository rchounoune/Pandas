### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Find the total number of players
total_players = len(purchase_data["SN"].unique())

# Creating Total Players Dataframe
pd.DataFrame({"Total Players": [total_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Making Calculations
unique_items = len(purchase_data["Item ID"].unique())

average_price = round(purchase_data["Price"].mean(), 2)

num_purchases = len(purchase_data["Purchase ID"])

total_revenue = purchase_data["Price"].sum()

# Summary Table of Analysis Dataframe
purchasing_analysis = pd.DataFrame(
    {"Number of Unique Items": [unique_items],
     "Average Price": [average_price],
     "Number of Purchases": [num_purchases],
     "Total Revenue": "$"+ str(total_revenue)
     }
)
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Taking care of duplicates
total_player_count = purchase_data.drop_duplicates(["SN"], keep="last")

# Gender count calcutions
male_count = (total_player_count["Gender"] == "Male").sum()
female_count = (total_player_count["Gender"] == "Female").sum()
others_count = (total_player_count["Gender"] == "Other / Non-Disclosed").sum()

# Percentage Calculations
percent_male = round((male_count/total_players*100),2)
percent_female = round((female_count/total_players*100),2)
percent_others = round((others_count/total_players*100),2)

# Creating summary of gender demographics 
gender_demographics = pd.DataFrame(
    {"":("Male","Female", "Other / Non-Disclosed"),
    "Total Count": (male_count,female_count,others_count),
     "Percentage of Players": (percent_male, percent_female, others_count),
    }
)

gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_gender = purchase_data.groupby("Gender").nunique()["SN"]
gender_count = purchase_data.groupby("Gender").count()["Price"]

avg_purch_price = round(purchase_data.groupby("Gender").mean()["Price"],2)

total_price = round(purchase_data.groupby("Gender").sum()["Price"],2)


avg_pers_price = round((total_price /total_gender),2)


purchasing_analysis = pd.DataFrame(
    {
    "Purchase Count" : gender_count,
    "Average Purchase Price": avg_purch_price,
    "Total Purchase Value" : total_price,
    "Avg Total Purchase per Person": avg_pers_price
}
)

purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Defininng Bins of ages and Ranges
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 44.99]
age_range = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=age_range)


# Calculate the numbers and percentages by age group

age = purchase_data.groupby("Age Range").nunique()["SN"]
percentage = round(age / total_players *100,2)

# Create a summary data frame to hold the results
age_analysis = pd.DataFrame({"Total Count": age,
             "Percentage of Players": percentage,
             })

age_analysis
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count,
# purch_count = purchase_data.groupby('Age Range').count()['Purchase ID']
purch_count = purchase_data.groupby('Age Range').count()['Purchase ID']
#avg. purchase price
avg_price = purchase_data.groupby('Age Range').mean()['Price']
avg_price = avg_price.map("${:.2f}".format)
#purchase total per person
total_price = purchase_data.groupby('Age Range').sum()['Price']

#avg. purchase total per person
avg_total_pp = total_price / age

avg_total_pp = avg_total_pp.map("${:.2f}".format)

purchase_data["avg_purchase_per_person"] = avg_total_pp

purchasing_analysis = pd.DataFrame({"Purchase Count": purch_count,
             "Average Purchase Price": avg_price,
             "Total Purchase Value": total_price,
             "Avg Total Purchase per Person": avg_total_pp
            }
            )

purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Spender Count,
spender_count = purchase_data.groupby('SN').nunique()['Purchase ID']

# avg. spender price
avg_spender_price = purchase_data.groupby('SN').mean()['Price']
avg_spender_price = avg_spender_price.map("${:.2f}".format)

# purchase total per spender
total_spender_price = purchase_data.groupby('SN').sum()['Price']
total_spender_price =total_spender_price.map("${:.2f}".format)

# Top spender Dataframe
top_spender = pd.DataFrame({"Purchase Count": spender_count,
             "Average Purchase Price": avg_spender_price,
             "Total Purchase Value": total_spender_price,
             })


top_spender.sort_values(by=['Total Purchase Value'], ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Establish new data
new_data = purchase_data[["Item ID","Item Name","Price"]]

popular_data = new_data.groupby(["Item ID", "Item Name"])

# Purchase Count
p_count = popular_data["Price"].count()

# Total price
total_price = popular_data["Price"].sum()

# Item price
item_price = total_price / p_count

# Popular Items Dataframe
popular_items = pd.DataFrame({"Purchase Count": p_count,
             "Item Price": total_price / p_count,
             "Total Purchase Value": popular_data["Price"].sum(),
             })


popular_items.sort_values(by=['Purchase Count'], ascending=False).head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order
